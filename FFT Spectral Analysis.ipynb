# FFT Spectral Analysis

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

import random
```

### 1、正弦波

```python
fs =100  ##设定采样频率
N = 128
n = np.arange(0,N,1)
t = n / fs
f0 = 10
x = np.sin(2 * np.pi * f0 * t)
```

使用matplotlib绘制谱图:

```python
fig,ax = plt.subplots()
ax.plot(t,x)
plt.xlabel('Time/s')
plt.ylabel('Amplitude')
plt.title('Time Domain Waveform')
plt.show()
```

进行FFT变换并绘制频谱图：

```python
y = np.fft.fft(x)
mag = abs(y)
f = np.arange(0,len(y),1)
f = f * fs / len(y)
```

```python
plt.plot(f,mag)
plt.xlabel('Frequency/s')
plt.ylabel('Amplitude')
plt.title('Frequency Waveform')
plt.show()
```

使用IFFT恢复原始信号：

```python
xifft = np.fft.ifft(y)
ti = np.arange(0,len(xifft),1)
ti = ti / fs
```

```python
plt.plot(ti,xifft)
plt.xlabel('Time/s')
plt.ylabel('Amplitude')
plt.title('IFFT Waveform')
plt.show()
```

## 2、白噪声

```python
fs = 50    #设定采样率
t = np.arange(-5,5,0.1)
x = np.zeros(100)
for i in range(100):
    x[i] = random.random()
```

```python
plt.plot(t,x)
plt.xlabel('Time/s')
plt.ylabel('Amplitude')
plt.title('Time Domain Waveform')
plt.show()
```

```python
y = np.fft.fft(x)
mag = abs(y)
f = np.arange(0,len(y),1)
f = f * fs / len(y)
```

```python
plt.plot(f,mag)
plt.xlabel('Frequency/s')
plt.ylabel('Amplitude')
plt.title('Frequency Waveform')
plt.show()
```


使用IFFT恢复原始信号：

```python
xifft = np.fft.ifft(y)
ti = np.arange(0,len(xifft),1)
ti = ti / fs
```

```python
plt.plot(ti,xifft)
plt.xlabel('Time/s')
plt.ylabel('Amplitude')
plt.title('IFFT Waveform')
plt.show()
```
